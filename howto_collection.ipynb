```python
## Summary
# prompt: Example of data collection; list, dict, set, and tuple to make clear the differences and common features.

# Data collection types: list, dict, set, tuple

# List: Ordered, mutable, allows duplicate members. Created with square brackets [].
my_list = [1, 2, 2, 'apple', 'banana']
print("List:", my_list)
print("Type:", type(my_list))
print("First element:", my_list[0]) # Access by index
my_list.append(3) # Modify
print("List after append:", my_list)

# Dictionary: Unordered (in Python versions before 3.7), mutable, indexed by keys, does not allow duplicate keys. Created with curly brackets {}.
my_dict = {'name': 'Alice', 'age': 30, 'city': 'New York'}
print("\nDictionary:", my_dict)
print("Type:", type(my_dict))
print("Name:", my_dict['name']) # Access by key
my_dict['age'] = 31 # Modify
my_dict['job'] = 'Engineer' # Add new key-value pair
print("Dictionary after modification:", my_dict)

# Set: Unordered, immutable (elements cannot be changed after creation, but items can be added or removed), does not allow duplicate members. Created with curly brackets {} or set().
my_set = {1, 2, 2, 'apple', 'banana'}
print("\nSet:", my_set) # Duplicate 2 is automatically removed
print("Type:", type(my_set))
my_set.add('orange') # Add element
my_set.remove(1) # Remove element
print("Set after modification:", my_set)

# Tuple: Ordered, immutable, allows duplicate members. Created with parentheses ().
my_tuple = (1, 2, 2, 'apple', 'banana')
print("\nTuple:", my_tuple)
print("Type:", type(my_tuple))
print("First element:", my_tuple[0]) # Access by index
# Tuples are immutable, the following line would raise an error:
# my_tuple[0] = 5
```

```python
# prompt: Detailed examples of ”list” especially w.r.t. the features; order, mutability, duplicate item.

# Detailed examples focusing on list features: order, mutability, duplicate items

# Creating a list with duplicate items
my_list = [10, 20, 30, 20, 40, 10, 50]
print("Original list:", my_list)

# --- Feature 1: Order ---
# Lists maintain the order in which elements are added.
# We can access elements by their index (position).
print("\nAccessing elements by index:")
print("First element:", my_list[0])
print("Third element:", my_list[2])
print("Last element:", my_list[-1])

# Slicing also respects order
print("Elements from index 1 to 3 (exclusive of 4):", my_list[1:4])

# The order is preserved even after modifications
my_list.append(60)
print("List after appending 60:", my_list)

# --- Feature 2: Mutability ---
# Lists are mutable, meaning their contents can be changed after creation.

# Changing an element at a specific index
my_list[1] = 25
print("\nList after changing element at index 1:", my_list)

# Adding elements (append, insert)
my_list.append(70)
my_list.insert(2, 35) # Insert 35 at index 2
print("List after appending 70 and inserting 35:", my_list)

# Removing elements (remove, pop, del)
my_list.remove(20) # Removes the first occurrence of 20
print("List after removing the first 20:", my_list)

popped_element = my_list.pop() # Removes and returns the last element
print("List after popping the last element:", my_list)
print("Popped element:", popped_element)

del my_list[0] # Deletes the element at index 0
print("List after deleting element at index 0:", my_list)

# --- Feature 3: Duplicate Items ---
# Lists allow multiple occurrences of the same value.

another_list = [1, 2, 2, 3, 1, 4, 2]
print("\nList with duplicate items:", another_list)

# Duplicates are treated as distinct elements based on their position.
print("Count of '2' in the list:", another_list.count(2))

# Removing a duplicate item using index will only remove one instance
del another_list[2] # Deletes the second '2'
print("List after deleting element at index 2:", another_list)

# Removing a duplicate item using remove() removes the first occurrence
another_list.remove(1)
print("List after removing the first '1':", another_list)

# Checking for membership works correctly with duplicates
print("Is 2 in the list?", 2 in another_list)
print("Is 5 in the list?", 5 in another_list)

# Sorting a list with duplicates keeps all duplicates
another_list.sort()
print("List after sorting:", another_list)
```

```python
# prompt: Detailed example of dict

# Detailed examples focusing on dictionary features: unordered (in older versions), mutable, indexed by keys, unique keys

# Creating a dictionary
my_dict = {
    'product_id': 'A123',
    'name': 'Laptop',
    'price': 1200.50,
    'in_stock': True,
    'tags': ['electronics', 'computer', 'sale'],
    'manufacturer': {'name': 'TechCorp', 'country': 'USA'} # Nested dictionary
}
print("Original dictionary:", my_dict)
print("Type:", type(my_dict))

# --- Feature 1: Unordered (in versions before Python 3.7), Ordered (from Python 3.7) ---
# Dictionaries do not maintain insertion order in older versions (< 3.7).
# From Python 3.7 onwards, dictionaries are ordered by insertion order.
# When printing or iterating, the order will match the insertion order.
a = {'one': 1, 'two': 2}
b = {'two': 2, 'one': 1}
cprint("a == b    ",f"{a == b}","   : matching of dict is element wise")
cprint("a[list(a.keys())[1]]    ",f"{a[list(a.keys())[1]]}","   :ordered by insertion order")

# --- Feature 2: Mutable ---
# Dictionaries are mutable, meaning their contents can be changed after creation.

# Adding a new key-value pair
my_dict['warranty'] = '1 year'
print("\nDictionary after adding 'warranty':", my_dict)

# Changing the value associated with an existing key
my_dict['price'] = 1150.00
print("Dictionary after changing 'price':", my_dict)

# Changing a value in a nested dictionary
my_dict['manufacturer']['country'] = 'Canada'
print("Dictionary after changing nested 'country':", my_dict)

# Removing a key-value pair
del my_dict['in_stock']
print("Dictionary after deleting 'in_stock':", my_dict)

# Using pop() to remove and get a value
removed_name = my_dict.pop('name')
print("Dictionary after popping 'name':", my_dict)
print("Removed name:", removed_name)

# --- Feature 3: Indexed by Keys ---
# Dictionary elements are accessed using their keys, not their index.

print("\nAccessing elements by key:")
print("Product ID:", my_dict['product_id'])
print("Price:", my_dict['price'])
print("Tags:", my_dict['tags'])
print("Manufacturer Name:", my_dict['manufacturer']['name']) # Accessing nested value

# Using .get() method for safe access (returns None if key doesn't exist, or a default value)
print("Warranty (using get()):", my_dict.get('warranty'))
print("Stock level (using get() with default):", my_dict.get('stock_level', 'Not specified'))
print("Non-existent key (using get()):", my_dict.get('color')) # Returns None

# Trying to access a non-existent key directly will raise a KeyError
# print(my_dict['color']) # This would cause an error

# --- Feature 4: Unique Keys ---
# Dictionary keys must be unique. If you try to add a key that already exists,
# the old value associated with that key will be overwritten.

duplicate_key_dict = {'a': 1, 'b': 2, 'a': 3}
print("\nDictionary with duplicate key 'a':", duplicate_key_dict) # The value 1 is overwritten by 3

# Attempting to add the same key again just updates the value
my_dict['price'] = 1100.00
print("Dictionary after setting 'price' again:", my_dict) # Value is updated

# Keys must be immutable types (strings, numbers, tuples). Lists or dictionaries cannot be keys.
# invalid_dict = {[1, 2]: 'list as key'} # This would cause a TypeError

# --- Dictionary Methods ---

print("\nDictionary Methods:")

# .keys() - returns a view object that displays a list of all the keys
print("Keys:", my_dict.keys())

# .values() - returns a view object that displays a list of all the values
print("Values:", my_dict.values())

# .items() - returns a view object that displays a list of key-value tuple pairs
print("Items:", my_dict.items())

# Iterating through a dictionary
print("\nIterating through items:")
for key, value in my_dict.items():
  print(f"{key}: {value}")

print("\nIterating through keys:")
for key in my_dict.keys(): # or just "for key in my_dict:"
  print(key)

print("\nIterating through values:")
for value in my_dict.values():
  print(value)

# Check if a key exists in the dictionary
print("\nChecking key existence:")
print("'product_id' in my_dict:", 'product_id' in my_dict)
print("'color' in my_dict:", 'color' in my_dict)

# Check if a value exists (less efficient for large dicts)
print("'Laptop' in my_dict.values():", 'Laptop' in my_dict.values())
print("'Tablet' in my_dict.values():", 'Tablet' in my_dict.values())

# Creating a dictionary using dict() constructor
dict_from_tuples = dict([('one', 1), ('two', 2)])
print("\nDict from tuples:", dict_from_tuples)

dict_from_keywords = dict(name='Charlie', age=25)
print("Dict from keywords:", dict_from_keywords)

# Copying a dictionary
copied_dict = my_dict.copy()
print("\nCopied dictionary:", copied_dict)
# Changes to copied_dict do not affect my_dict (for top level keys)
copied_dict['price'] = 500
print("Original dict price after changing copy:", my_dict['price'])
print("Copied dict price:", copied_dict['price'])

# However, changes to mutable values within a copy (like a list or nested dict)
# *will* affect the original if a shallow copy (.copy()) is used.
# For deep copies, use the 'copy' module: import copy; deep_copied_dict = copy.deepcopy(my_dict)

# Clearing a dictionary
# my_dict.clear()
# print("\nDictionary after clearing:", my_dict)

```

```python
# prompt: Detailed example of set

# Detailed examples focusing on set features: unordered, mutable (elements can be added/removed, but not changed), no duplicate members

# Creating a set
my_set = {10, 20, 30, 20, 40, 10, 50} # Duplicates 20 and 10 are automatically removed
print("Original set:", my_set)
print("Type:", type(my_set))

# Creating an empty set (use set(), not {}) - {} creates an empty dictionary
empty_set = set()
print("Empty set:", empty_set)
print("Type of {} is:", type({}))

# --- Feature 1: Unordered ---
# Sets do not maintain any specific order. The order of elements when you print a set
# might not be the same as the order in which they were added, and can vary.
# You cannot access elements by index or slice sets.
# print(my_set[0]) # This would cause a TypeError

# --- Feature 2: Mutable (Add/Remove items, but elements themselves are immutable) ---
# Sets are mutable, meaning elements can be added or removed.
# However, the elements themselves must be immutable types (like numbers, strings, tuples).
# Lists, dictionaries, or other sets cannot be direct members of a set because they are mutable.

# Adding elements
my_set.add(60)
print("\nSet after adding 60:", my_set)

my_set.add(30) # Adding an existing element has no effect
print("Set after adding 30 (already exists):", my_set)

# Adding multiple elements using update()
my_set.update([70, 80])
print("Set after updating with [70, 80]:", my_set)

my_set.update({90, 100}) # update() can also take other iterables like sets or tuples
print("Set after updating with {90, 100}:", my_set)

# Removing elements
my_set.remove(10) # Removes the specified element
print("Set after removing 10:", my_set)

# my_set.remove(99) # Removing a non-existent element with remove() raises a KeyError

my_set.discard(20) # Removes the specified element if it is present
print("Set after discarding 20:", my_set)

my_set.discard(99) # Removing a non-existent element with discard() does NOT raise an error
print("Set after discarding 99 (non-existent):", my_set)

popped_element = my_set.pop() # Removes and returns an arbitrary element (since sets are unordered)
print("Set after popping an element:", my_set)
print("Popped element:", popped_element)

# Clearing a set
# my_set.clear()
# print("Set after clearing:", my_set)

# --- Feature 3: No Duplicate Members ---
# Sets automatically eliminate duplicate values. Only unique elements are stored.

duplicate_set = {1, 2, 2, 3, 1, 4, 2}
print("\nSet created with duplicates:", duplicate_set) # Output will only show unique elements {1, 2, 3, 4}

# --- Set Operations ---
# Sets are powerful for mathematical set operations.

set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

print("\nSet Operations:")

# Union: elements in either set1 or set2 (or both)
union_set = set1.union(set2)
print("Union (set1 | set2):", union_set)
# Alternative syntax using | operator
print("Union (set1 | set2) using operator:", set1 | set2)

# Intersection: elements common to both set1 and set2
intersection_set = set1.intersection(set2)
print("Intersection (set1 & set2):", intersection_set)
# Alternative syntax using & operator
print("Intersection (set1 & set2) using operator:", set1 & set2)

# Difference: elements in set1 but not in set2
difference_set1_2 = set1.difference(set2)
print("Difference (set1 - set2):", difference_set1_2)
# Alternative syntax using - operator
print("Difference (set1 - set2) using operator:", set1 - set2)

# Difference: elements in set2 but not in set1
difference_set2_1 = set2.difference(set1)
print("Difference (set2 - set1):", difference_set2_1)
# Alternative syntax using - operator
print("Difference (set2 - set1) using operator:", set2 - set1)


# Symmetric Difference: elements in either set1 or set2, but not in both
symmetric_difference_set = set1.symmetric_difference(set2)
print("Symmetric Difference (set1 ^ set2):", symmetric_difference_set)
# Alternative syntax using ^ operator
print("Symmetric Difference (set1 ^ set2) using operator:", set1 ^ set2)

# --- Set Relationships ---

set_a = {1, 2, 3}
set_b = {1, 2, 3, 4, 5}
set_c = {1, 2, 6}

print("\nSet Relationships:")

# Subset: True if all elements of set_a are in set_b
print("set_a is subset of set_b:", set_a.issubset(set_b))
# Alternative syntax using <= operator
print("set_a <= set_b:", set_a <= set_b)
print("set_b is subset of set_a:", set_b.issubset(set_a)) # False

# Proper Subset: True if set_a is a subset of set_b and set_a is not equal to set_b
print("set_a is proper subset of set_b:", set_a < set_b)
print("set_a is proper subset of set_a:", set_a < set_a) # False

# Superset: True if all elements of set_b are in set_a
print("set_b is superset of set_a:", set_b.issuperset(set_a))
# Alternative syntax using >= operator
print("set_b >= set_a:", set_b >= set_a)
print("set_a is superset of set_b:", set_a.issuperset(set_b)) # False

# Proper Superset: True if set_a is a superset of set_b and set_a is not equal to set_b
print("set_b is proper superset of set_a:", set_b > set_a)
print("set_a is proper superset of set_a:", set_a > set_a) # False


# Disjoint: True if the sets have no elements in common (their intersection is an empty set)
print("set_a is disjoint with set_b:", set_a.isdisjoint(set_b)) # False (they share 1, 2, 3)
print("set_a is disjoint with set_c:", set_a.isdisjoint(set_c)) # False (they share 1, 2)
set_d = {7, 8, 9}
print("set_a is disjoint with set_d:", set_a.isdisjoint(set_d)) # True

# --- Other Set Features ---

# Length of a set
print("\nLength of set_b:", len(set_b))

# Membership testing (very efficient)
print("3 in set_a:", 3 in set_a)
print("6 in set_a:", 6 in set_a)

# Iterating through a set (order is not guaranteed)
print("\nIterating through set_b:")
for element in set_b:
  print(element)

# Frozen Sets: Immutable Sets
# Frozen sets are like sets but are immutable. They can be used as elements in another set
# or as keys in a dictionary because they are hashable.
frozen_set = frozenset([1, 2, 3, 2])
print("\nFrozen Set:", frozen_set)
print("Type:", type(frozen_set))

# frozen_set.add(4) # This would cause a TypeError
# frozen_set.remove(1) # This would cause a TypeError

# Using a frozen set as a dictionary key
my_complex_dict = {
    frozenset({'a', 'b'}): 'value 1',
    frozenset({'c', 'd'}): 'value 2'
}
print("\nDictionary with frozenset keys:", my_complex_dict)

# Checking membership with a frozenset key
print("Membership check:", frozenset({'b', 'a'}) in my_complex_dict) # Order doesn't matter for set keys
print("Membership check:", frozenset({'a', 'c'}) in my_complex_dict) # False
```

```python
# prompt: Detailed example of tuple

# Detailed examples focusing on tuple features: ordered, immutable, allows duplicate members

# Creating a tuple
my_tuple = (10, 20, 30, 20, 40, 10, 50)
print("Original tuple:", my_tuple)
print("Type:", type(my_tuple))

# Creating a tuple with mixed data types
mixed_tuple = (1, "hello", 3.14, [1, 2], {'a': 1})
print("\nMixed data type tuple:", mixed_tuple)

# Creating a tuple with a single element (requires a trailing comma)
single_element_tuple = (5,)
print("\nSingle element tuple:", single_element_tuple)
print("Type:", type(single_element_tuple))

# Without the comma, it's just an expression in parentheses
not_a_tuple = (5)
print("Not a tuple:", not_a_tuple)
print("Type:", type(not_a_tuple))

# Creating a tuple from an iterable (like a list)
tuple_from_list = tuple([1, 2, 3, 4])
print("\nTuple from list:", tuple_from_list)

# --- Feature 1: Ordered ---
# Tuples maintain the order in which elements are added.
# We can access elements by their index (position).
print("\nAccessing elements by index:")
print("First element:", my_tuple[0])
print("Third element:", my_tuple[2])
print("Last element:", my_tuple[-1])

# Slicing also respects order
print("Elements from index 1 to 3 (exclusive of 4):", my_tuple[1:4])

# --- Feature 2: Immutable ---
# Tuples are immutable, meaning their contents cannot be changed after creation.

# Attempting to change an element will raise a TypeError
# my_tuple[1] = 25 # This line would cause a TypeError: 'tuple' object does not support item assignment

# Attempting to add or remove elements will also raise errors
# my_tuple.append(60) # This line would cause a TypeError: 'tuple' object has no attribute 'append'
# my_tuple.remove(10) # This line would cause a TypeError: 'tuple' object has no attribute 'remove'

# Although a tuple itself is immutable, if it contains mutable objects (like lists or dictionaries),
# the contents of those mutable objects *can* be changed.
mutable_in_tuple = ([1, 2], {'a': 1})
print("\nTuple with mutable element:", mutable_in_tuple)
mutable_in_tuple[0].append(3) # Modifying the list inside the tuple
print("Tuple after modifying inner list:", mutable_in_tuple)
mutable_in_tuple[1]['b'] = 2 # Modifying the dictionary inside the tuple
print("Tuple after modifying inner dictionary:", mutable_in_tuple)
# mutable_in_tuple[0] = [4,5] # This would still raise a TypeError because we're trying to assign a new list to index 0 of the tuple

# To "modify" a tuple, you typically create a new tuple based on the original.
new_tuple = my_tuple[:2] + (99,) + my_tuple[3:]
print("\nOriginal tuple (unchanged):", my_tuple)
print("New tuple created from original:", new_tuple)

# --- Feature 3: Allows Duplicate Members ---
# Tuples allow multiple occurrences of the same value.

duplicate_tuple = (1, 2, 2, 3, 1, 4, 2)
print("\nTuple with duplicate items:", duplicate_tuple)

# Duplicates are treated as distinct elements based on their position.
print("Count of '2' in the tuple:", duplicate_tuple.count(2))

# --- Tuple Methods ---

print("\nTuple Methods:")

# .count(value) - returns the number of times a specified value occurs in the tuple
print("Count of 20 in my_tuple:", my_tuple.count(20))
print("Count of 99 in my_tuple:", my_tuple.count(99))

# .index(value) - searches the tuple for a specified value and returns the position of where it was found
# If the value is present more than once, it returns the index of the first occurrence.
print("Index of the first 30 in my_tuple:", my_tuple.index(30))
# print(my_tuple.index(99)) # This would cause a ValueError if 99 is not in the tuple

# --- Other Tuple Features ---

# Length of a tuple
print("\nLength of my_tuple:", len(my_tuple))

# Membership testing
print("30 in my_tuple:", 30 in my_tuple)
print("99 in my_tuple:", 99 in my_tuple)

# Iterating through a tuple (order is guaranteed)
print("\nIterating through my_tuple:")
for element in my_tuple:
  print(element)

# Tuple Unpacking: Assigning elements of a tuple to multiple variables
coordinates = (10, 20)
x, y = coordinates
print(f"\nUnpacked coordinates: x={x}, y={y}")

person = ("Alice", 30, "New York")
name, age, city = person
print(f"Unpacked person: Name={name}, Age={age}, City={city}")

# Swapping variables using tuple unpacking
a = 5
b = 10
print(f"\nBefore swap: a={a}, b={b}")
a, b = b, a
print(f"After swap: a={a}, b={b}")

# Returning multiple values from a function (functions often return tuples)
def get_minmax(numbers):
  if not numbers:
    return None, None # Return a tuple of None values
  return min(numbers), max(numbers) # Return a tuple

numbers = [10, 5, 25, 15]
minimum, maximum = get_minmax(numbers)
print(f"\nMin and max of {numbers}: min={minimum}, max={maximum}")

empty_list = []
minimum, maximum = get_minmax(empty_list)
print(f"Min and max of {empty_list}: min={minimum}, max={maximum}")

# Tuples are often used for heterogeneous data (different data types)
# Lists are often used for homogeneous data (same data types)

# Tuples are immutable and thus hashable, making them suitable for use as dictionary keys
# or elements in sets, unlike lists.
my_dict_with_tuple_key = {
    (1, 2): 'value for (1, 2)',
    (3, 4): 'value for (3, 4)'
}
print("\nDictionary with tuple keys:", my_dict_with_tuple_key)
print("Value for key (1, 2):", my_dict_with_tuple_key[(1, 2)])

# my_dict_with_list_key = {[1, 2]: 'value'} # This would cause a TypeError

my_set_with_tuples = {(1, 2), (3, 4), (1, 2)} # Duplicate tuple (1, 2) is removed
print("\nSet with tuples:", my_set_with_tuples)

# my_set_with_lists = {[1, 2], [3, 4]} # This would cause a TypeError
```
